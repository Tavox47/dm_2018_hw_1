```python
# Answer here
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

### Reduction
X_3D_reduced = PCA(n_components = 3).fit_transform(X_counts.toarray())
### 3D plotting
col = ['coral', 'blue', 'black', 'm']
# plot
fig = plt.figure(figsize = (25,10))
ax = fig.add_subplot(221, projection='3d')
bx = fig.add_subplot(222, projection='3d')
cx = fig.add_subplot(223, projection='3d')

for c, category in zip(col, categories):
    xs = X_3D_reduced[X['category_name'] == category].T[0]
    ys = X_3D_reduced[X['category_name'] == category].T[1]
    zs = X_3D_reduced[X['category_name'] == category].T[2]
    
    ax.scatter(xs, ys, zs, c = c, marker='o')
    ax.view_init(45,45)
    bx.scatter(xs, ys, zs, c = c, marker='o')
    bx.view_init(45,0)
    cx.scatter(xs, ys, zs, c = c, marker='o')

ax.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
ax.set_xlabel('\nX Label')
ax.set_ylabel('\nY Label')
ax.set_zlabel('\nZ Label')

bx.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
bx.set_xlabel('\nX Label')
bx.set_ylabel('\nY Label')
bx.set_zlabel('\nZ Label')

cx.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
cx.set_xlabel('\nX Label')
cx.set_ylabel('\nY Label')
cx.set_zlabel('\nZ Label')

plt.show()
```

```python
#Create TFIDF Matrix
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
TfIdf = vectorizer.fit_transform(dataset.sentence)
#print(vectorizer.get_feature_names())
print(TfIdf.shape)
TfIdf[0:5, 0:60].toarray()
```

```python
#Create Freq Matrix
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
Freq = count_vect.fit_transform(dataset.sentence)
print(Freq.shape)
Freq[0:5, 0:60].toarray()
```

Can notice in the 4th tupple there is one non-zero entry. In TfIdf it is the score 0.35888007 while in Freq is only the count 1.

```python
#Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity

#Test case 1
cosim1_tfidf = cosine_similarity(TfIdf[0:1],TfIdf[100:101])
cosim1_freq = cosine_similarity(Freq[0:1],Freq[100:101])

#Test case 2
cosim2_tfidf = cosine_similarity(TfIdf[200:201],TfIdf[1000:1001])
cosim2_freq = cosine_similarity(Freq[200:201],Freq[1000:1001])

print(cosim1_tfidf, cosim1_freq)
print(cosim2_tfidf, cosim2_freq)
```

```python
#Manhattan Distance
from sklearn.metrics.pairwise import manhattan_distances

#Test case 1
mandis1_tfidf = manhattan_distances(TfIdf[0:1],TfIdf[100:101])
mandis1_freq = manhattan_distances(Freq[0:1],Freq[100:101])

#Test case 2
mandis2_tfidf = manhattan_distances(TfIdf[200:201],TfIdf[1000:1001])
mandis2_freq = manhattan_distances(Freq[200:201],Freq[1000:1001])

print(mandis1_tfidf, mandis1_freq)
print(mandis2_tfidf, mandis2_freq)
```

Tried Cosine Similarity and Manhattan distance for 2 pairs of randomly selected sentences. First thing to notice with both cosine similarity and manhattan distance is that the scores when using Tf-Idf as features will be smaller because the weights are smaller than the counts used in Frequency. However when comparing between case1 and case2 we can clearly see that the sentences in case 2 are "more similar". 
For the case of Manhattan Distances we must remember the value is a distance(dissimilarity) so it must me interpreted reversely, the lower the value, the more similar they are.
In the selected examples there is consistency in results accross similarity measures as well as feature used.

```python
#Naive Bayes Classifier(Multinomial)
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

dataset['score'].replace(["negative"],'0',inplace=True)
dataset['score'].replace(["positive"],'1',inplace=True)
Y = np.array(dataset.score)
Y
```

```python
#With TfIdf features
import matplotlib.pyplot as plt

X_train, X_test, Y_train, Y_test = train_test_split(TfIdf, Y, test_size=0.20)

mnb = MultinomialNB()
mnb.fit(X_train, Y_train)
Y_mnb_score = mnb.predict_proba(X_test)

print(mnb.score(X_test, Y_test))
fpr_mnb, tpr_mnb, thresholds_mnb = roc_curve(Y_test, Y_mnb_score[:,1], pos_label='1') 
print (auc(fpr_mnb, tpr_mnb))
plt.plot(fpr_mnb,tpr_mnb)
plt.show()
```

```python
#With Frequency features
import matplotlib.pyplot as plt

X_train, X_test, Y_train, Y_test = train_test_split(Freq, Y, test_size=0.20)

mnb = MultinomialNB()
mnb.fit(X_train, Y_train)
Y_mnb_score = mnb.predict_proba(X_test)

print(mnb.score(X_test, Y_test))
fpr_mnb, tpr_mnb, thresholds_mnb = roc_curve(Y_test, Y_mnb_score[:,1], pos_label='1') 
print (auc(fpr_mnb, tpr_mnb))
plt.plot(fpr_mnb,tpr_mnb)
plt.show()
```

```python
#Logistic Regression for comparison
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#With TfIdf features
X_train, X_test, Y_train, Y_test = train_test_split(TfIdf, Y, test_size=0.20)

lr = LogisticRegression()
lr.fit(X_train, Y_train)
Y_lr_score = lr.decision_function(X_test)

print(mnb.score(X_test, Y_test))
fpr_lr, tpr_lr, thresholds_lr = roc_curve(Y_test, Y_lr_score, pos_label='1') 
print (auc(fpr_lr, tpr_lr))
plt.plot(fpr_lr,tpr_lr)
plt.show()
```

```python
#With Freq features
X_train, X_test, Y_train, Y_test = train_test_split(Freq, Y, test_size=0.20)

lr = LogisticRegression()
lr.fit(X_train, Y_train)
Y_lr_score = lr.decision_function(X_test)

print(mnb.score(X_test, Y_test))
fpr_lr, tpr_lr, thresholds_lr = roc_curve(Y_test, Y_lr_score, pos_label='1') 
print (auc(fpr_lr, tpr_lr))
plt.plot(fpr_lr,tpr_lr)
plt.show()
```

Frequency feature gives a slightly better accuracy for Naive Bayes classifier. This is perhaps due to the nature of Bayesian probability which is directly based on frequency counts. In contrast we also train a logistic regression model, which we can assume makes better use of the Tf-Idf feature and provides better performance in terms of accuracy and auc(compared to Frequency). It is also interesting to see that accuracy us better in Logistic Regression for both features, but AUC is higher in the NB classifier.
